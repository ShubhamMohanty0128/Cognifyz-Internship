 Task 1: Top Cuisines
 Determine the top three most
 common cuisines in the dataset.
 Calculate
the
percentage
of
 restaurants that serve each of the top
 cuisines

```python
import pandas as pd

restaurants_df = pd.read_csv('/content/Dataset .csv')

correct_column_name = 'Cuisines'

cuisine_counts = restaurants_df[correct_column_name].value_counts()

top_cuisines = cuisine_counts.head(3)

total_restaurants = len(restaurants_df)
top_cuisines_percentage = (top_cuisines / total_restaurants) * 100

print("Top Three Cuisines:")
print(top_cuisines)
print("\nPercentage of Restaurants Serving Each Top Cuisine:")
print(top_cuisines_percentage)
```

Task 2: City Analysis
 Identify the city with the highest number
 of restaurants in the dataset.
 Calculate
the average rating for
 restaurants in each city.
 Determine the city with the highest
 average rating

```python
import pandas as pd

restaurants_df = pd.read_csv('/content/Dataset .csv')

city_column_name = 'City'

city_count = restaurants_df[city_column_name].value_counts()
highest_restaurant_city = city_count.idxmax()

average_ratings = restaurants_df.groupby(city_column_name)['Aggregate rating'].mean()

highest_rating_city = average_ratings.idxmax()
highest_average_rating = average_ratings.max()

print(f"The city with the highest number of restaurants: {highest_restaurant_city} with {city_count[highest_restaurant_city]} restaurants")
print(f"The city with the highest average rating: {highest_rating_city} with an average rating of {highest_average_rating:.2f}")
```

Task 3: Price Range Distribution
 Create a histogram or bar chart to
 visualize the distribution of price ranges
 among the restaurants.
 Calculate the percentage of restaurants
 in each price range category

```python
import pandas as pd
import matplotlib.pyplot as plt

restaurants_df = pd.read_csv('/content/Dataset .csv')

price_range_column = 'Price range'

price_range_counts = restaurants_df[price_range_column].value_counts()

total_restaurants = len(restaurants_df)
price_range_percentages = (price_range_counts / total_restaurants) * 100

plt.figure(figsize=(10, 6))
plt.bar(price_range_percentages.index, price_range_percentages.values)
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

for i, v in enumerate(price_range_percentages.values):
plt.text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom')

plt.show()

print("Percentage of Restaurants in Each Price Range Category:")
for price_range, percentage in price_range_percentages.items():
print(f"Price Range {price_range}: {percentage:.1f}%")
```

Task 4: Online Delivery
 Determine the percentage of restaurants
 that offer online delivery.
 Compare the average ratings of restaurants
 with and without online delivery

```python
import pandas as pd

restaurants_df = pd.read_csv('/content/Dataset .csv')

online_delivery_column = 'Has Online delivery'

online_delivery_restaurants = restaurants_df[restaurants_df[online_delivery_column] == 'Yes']
percentage_online_delivery = (len(online_delivery_restaurants) / len(restaurants_df)) * 100

average_rating_online_delivery = online_delivery_restaurants['Aggregate rating'].mean()
average_rating_no_online_delivery = restaurants_df[restaurants_df[online_delivery_column] == 'No']['Aggregate rating'].mean()

print(f"Percentage of restaurants offering online delivery: {percentage_online_delivery:.1f}%")
print(f"Average rating of restaurants with online delivery: {average_rating_online_delivery:.2f}")
print(f"Average rating of restaurants without online delivery: {average_rating_no_online_delivery:.2f}")
```

Task 2: Cuisine Combination

 cuisines in the dataset.

 tend to have higher ratings

```python
import pandas as pd
from collections import Counter

restaurants_df = pd.read_csv('/content/Dataset .csv')

cuisine_combinations = restaurants_df['Cuisines'].str.split(', ')
cuisine_combinations = cuisine_combinations.dropna()
all_combinations = [tuple(sorted(c)) for c in cuisine_combinations]
combination_counts = Counter(all_combinations)
most_common_combinations = combination_counts.most_common(10)

print("Most Common Cuisine Combinations:")
for combination, count in most_common_combinations:
    print(f"{combination}: {count}")

target_combinations = [('North Indian', 'Chinese'), ('Cafe', 'North Indian')]
```

Task 3: Geographic Analysis
 Plot the locations of restaurants on a
 map using longitude and latitude
 coordinates.
 Identify any patterns or clusters of
 restaurants in specific areas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

restaurants_df = pd.read_csv('/content/Dataset .csv')

longitude = restaurants_df['Longitude']
latitude = restaurants_df['Latitude']

plt.figure(figsize=(10, 6))
sns.scatterplot(x=longitude, y=latitude, data=restaurants_df)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Locations')
plt.show()

from sklearn.cluster import DBSCAN
import numpy as np

coordinates = np.column_stack([longitude, latitude])
db = DBSCAN(eps=0.005, min_samples=5).fit(coordinates)
labels = db.labels_
restaurants_df['Cluster'] = labels

plt.figure(figsize=(10, 6))
sns.scatterplot(x=longitude, y=latitude, hue='Cluster', data=restaurants_df, palette='viridis')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Clusters')
plt.show()
```

Task 4: Restaurant Chains
 Identify if there are any restaurant chains
 present in the dataset.
 Analyze the ratings and popularity of
 different restaurant chains

```python
import pandas as pd

restaurants_df = pd.read_csv('/content/Dataset .csv')

restaurant_counts = restaurants_df['Restaurant Name'].value_counts()
restaurant_chains = restaurant_counts[restaurant_counts > 1].index.tolist()

print("Restaurant Chains:")
print(restaurant_chains)

chain_ratings = {}
chain_popularity = {}

for chain in restaurant_chains:
    chain_data = restaurants_df[restaurants_df['Restaurant Name'] == chain]
    chain_ratings[chain] = chain_data['Aggregate rating'].mean()
    chain_popularity[chain] = len(chain_data)

print("\nChain Ratings:")
for chain, rating in chain_ratings.items():
    print(f"{chain}: {rating:.2f}")

print("\nChain Popularity (Number of Outlets):")
for chain, popularity in chain_popularity.items():
    print(f"{chain}: {popularity}")
```
