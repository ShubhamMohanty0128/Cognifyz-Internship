```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt #visualizing data 
%matplotlib inline
import seaborn as sns
```

```python
restaurants_df = pd.read_csv(r"C:\Users\Shubham\Downloads\cognifyz\Dataset .csv",encoding= 'unicode_escape')
```

```python
if 'Latitude' in restaurants_df.columns and 'Longitude' in restaurants_df.columns:
    # Drop rows with missing latitude or longitude
    restaurants_df = restaurants_df.dropna(subset=['Latitude', 'Longitude'])
    # Set up the plot
    plt.figure(figsize=(10, 8))
    sns.scatterplot(
        x='Longitude',
        y='Latitude',
        hue='Aggregate rating',
        size='Votes', 
        sizes=(20, 200),  
        data=restaurants_df,
        palette='viridis',
        alpha=0.7
    )
    # plot details
    plt.title('Restaurant Locations by Rating and Votes', fontsize=16)
    plt.xlabel('Longitude', fontsize=12)
    plt.ylabel('Latitude', fontsize=12)
    plt.legend(title='Rating', loc='upper right')
    plt.grid(True)
    plt.show()
else:
    print("The dataset does not contain 'Latitude' and 'Longitude' columns for geographic analysis.")
```
